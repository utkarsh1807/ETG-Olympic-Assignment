# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('summer.csv')

len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data=[]
da=df['Medal'].values.tolist()
ds=df['Sport'].values.tolist()
for i in range(len(da)):
    if(da[i]=='Gold'):
        data.append(ds[i])
    
data=pd.DataFrame(data,columns=['Sports'])

data1=[]

for sport in data['Sports'].unique():
    data1.append([sport,len(data[data['Sports']==sport])])
    
data1=pd.DataFrame(data1,columns=['Sports','No. of Gold Medals']).sort_values(by='No. of Gold Medals',ascending=False)

data1.head().plot(x='Sports',y='No. of Gold Medals',kind='bar')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]

for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
    
data=pd.DataFrame(data,columns=['Sports','Medals'])
data.sort_values(by='Medals',ascending=False).head().plot(x='Sports',y='Medals',kind='bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
d=df['Athlete'].unique().tolist()
d=pd.DataFrame(d,columns=['Player'])
d1=d['Player'].tolist()

p=[]

for x in range(len(d1)):
    p.append([d1[x],len(df[df['Athlete']==d1[x]])])
    
p=pd.DataFrame(p,columns=['Player','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Player',y='Medals',kind='bar')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data=[]

da=df['Medal'].values.tolist()
ds=df['Athlete'].values.tolist()
for i in range(len(da)):
    if(da[i]=='Gold'):
        data.append(ds[i])
        
data=pd.DataFrame(data,columns=['Player'])

g=[]

for player in data['Player'].unique():
    g.append([player,len(df[df['Athlete']==player])])

g=pd.DataFrame(g,columns=['Players','Gold Medals']).sort_values(by='Gold Medals',ascending=False).head().plot(x='Players',y='Gold Medals',kind='bar')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
c=df['Country'].tolist()
y=df['Year'].tolist()
m=df['Medal'].tolist()

z=0
for i in range(len(c)):
    if(c[i]=='IND' and m[i]=='Gold'):
        z=y[i]
        break
print(z)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]

for event in df['Event'].unique():
  data.append([event,len(df[df['Event']==event])])

data=pd.DataFrame(data,columns=['Event','freq'])

data=data.sort_values(by='freq',ascending=False).head().plot(x='Event',y='freq',kind='bar')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
da=df['Medal'].values.tolist()
ds=df['Sport'].values.tolist()
gen=df['Gender'].values.tolist()
for i in range(len(da)):
    if(da[i]=='Gold' and gen[i]=='Women'):
        data.append(ds[i])
        
data=pd.DataFrame(data,columns=['Sports'])

data1=[]

for sport in data['Sports'].unique():
    data1.append([sport,len(data[data['Sports']==sport])])
    
data1=pd.DataFrame(data1,columns=['Sports','No. of Female Gold Medalists']).head()

data1.sort_values(by='No. of Female Gold Medalists',ascending=False).plot(x='Sports',y='No. of Female Gold Medalists',kind='bar')
```
